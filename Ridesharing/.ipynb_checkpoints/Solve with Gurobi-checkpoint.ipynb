```python
import copy
import numpy as np
import igraph
from igraph import *

class VPG(object):
    def __init__(self,GraphFile,VPFile):
        self.graphfile = GraphFile
        self.VPFile = VPFile
        self.VPR = []
        self.PR = []
    
    def readFile(self):
        with open(self.graphfile) as graphfile:
            self.PR = [[int(num) for num in line.split(' ')] for line in graphfile]

        with open(self.VPFile) as VPFile:
            self.VPR = [[int(num) for num in line.split(' ')] for line in VPFile]
            
    
    def PPRelation(self):
        G = dict()
        for i in range(len(self.PR)):
            G[i] = []
            for j in range(len(self.PR)):
                if M[i][j] > 0:
                    G[i] += [j]
        self.PR = G
        return self.PR
    
    def VPRelation(self):
        G = dict()
        for i in range(len(self.VPR)):
            G[i] = []
            for j in range(len(self.VPR[0])):
                if VPR[i][j] > 0:
                    G[i] += [j]
        self.VPR = G
        return self.VPR
    
    def VehicleTripMatch(self,vehicle,Trip):
        if sum([self.VPR[vehicle][trip] == 1 for trip in Trip]) == len(Trip):
            return True
        else:
            return False
    
    def ConstructGraph(self):
        return {'Vehicle-Passenger':self.VPR,'Passenger-Passenger':self.PR}
    
    def Vehicles(self):
        if len(self.VPR) == 0:
            return 0
        else:
            return [i for i in range(len(self.VPR))]
    
    def Passengers(self):
        if len(self.VPR[0]) == 0:
            return 0
        else:
            return [i for i in range(len(self.PR))]
        
class trip(object):
    def __init__(self,passengers):
        self.passengers = passengers
        
class VTG(VPG):
    def __init__(self,GraphFile,VPFile,capacity):
        vpg = VPG(GraphFile,VPFile)
        vpg.readFile()
        self.V = vpg.Vehicles()
        self.P = vpg.Passengers()
        self.VPR = vpg.VPR
        self.PPR = vpg.PR
        self.VPG = vpg
        self.max_capacity = capacity
        self.neighbors = dict()
        self.Graph = []
        self.VehicleTripMatrix = []
        self.TripTripMatrix = []
        self.VehicleTripGraph = igraph.Graph()
        self.TripTripGraph = igraph.Graph()
        self.num_vehicles = len(self.V)
        self.num_trips = 0
        
    def VTG(self):
        Trip = [] 
        for i in range(self.max_capacity):
            Trip.append([])
            
        # add trips with one passenger into Trip
        for p in self.P:     
            Trip[0].append((p,))
        
        # add trips with two passengers into Trip
        for i in range(len(Trip[0])-1):
            v_1 = Trip[0][i][0]
            
            for j in range(i+1,len(Trip[0])):
                v_2 = Trip[0][j][0]

                if self.PPR[v_1][v_2] == 1: 
                    Trip[1].append((v_1,v_2))
                    
            
        for k in range(2,self.max_capacity):
            len_TTM = len(self.TripTripMatrix)
            for i in range(len(Trip[k-1])):
                T_1 = Trip[k-1][i]
                
                for j in range(i+1,len(Trip[k-1])):
                    T_2 = Trip[k-1][j]
                    T_set = set(list(T_1)+list(T_2))
                    
                        
                    if len(T_set) == (k+1):
                        flag = 0
                        for r in range(k+1):
                            T = tuple(T_set)
                            T_list = list(T_set)
                            T_set_copy = copy.copy(T_set)
                            T_set_copy.remove(T_list[r])
                            
                            if tuple(T_set_copy) in Trip[k-1]:
                                flag += 1
                        if (flag == len(T_list)) & (T not in Trip[k]):
                            Trip[k].append(T)
        
        Trips = []
        for i in range(len(Trip)):
            Trips += Trip[i]   
        self.Trips = Trips
        self.num_trips = len(Trips)
#         self.VehicleTripGraph.add_vertices(self.num_vehicles+self.num_trips)
#         self.TripTripGraph.add_vertices(self.num_trips)
        
        VTAdjacencyMatrix = []
        self.VTAdjacencyMatrix = VTAdjacencyMatrix
        for v in self.V:
#             self.VehicleTripGraph.vs[v]['id'] = v
            num = self.num_vehicles
            VTMatrix = []
            for i in range(len(Trip[0])):
                v_1 = Trip[0][i][0]
#                 self.VehicleTripGraph.vs[i+num]['id'] = i+num
#                 self.VehicleTripGraph.vs[i+num]['num_passengers'] = 1
                if (self.VPR[v][v_1] == 1):
#                     self.VehicleTripGraph.add_edge(v,i+self.num_vehicles)
                    VTMatrix.append(1)
                else:
                    VTMatrix.append(0)
            num += len(Trip[0])
            
            
            for i in range(len(Trip[1])):
                (v_1,v_2) = Trip[1][i]
#                 self.VehicleTripGraph.vs[i+num]['id'] = i+num
#                 self.VehicleTripGraph.vs[i+num]['num_passengers'] = 2
                if (self.VPR[v][v_1] == 1) and (self.VPR[v][v_2] == 1):
#                     self.VehicleTripGraph.add_edge(v,i+num)
                    VTMatrix.append(1)
                else:
                    VTMatrix.append(0)
            num += len(Trip[1])
                
            for j in range(2,len(Trip)):
                for i in range(len(Trip[j])):
                    T_list = Trip[j][i]
                    
#                     self.VehicleTripGraph.vs[i+num]['id'] = i+num
#                     self.VehicleTripGraph.vs[i+num]['num_passengers'] = j+1
                    if self.VPG.VehicleTripMatch(v,T_list):
#                         self.VehicleTripGraph.add_edge(v,i+num)
                        VTMatrix.append(1)
                    else:
                        VTMatrix.append(0)
                num += len(Trip[j])
            VTAdjacencyMatrix.append(VTMatrix)
            
        num = self.num_vehicles                                       
#         for i in range(len(Trips)):
#             self.TripTripGraph.vs[i]['id'] = i + num
#             self.TripTripGraph.vs[i]['ind'] = i
#             self.TripTripGraph.vs[i]['num_passengers'] = len(Trips[i])
#             for k in range(i+1,len(Trips)):
#                 if (set(Trips[i]) & set(Trips[k])) != set():
#                     self.TripTripGraph.add_edge(i,k)
    
        
#         self.num_v = len(self.VehicleTripGraph.vs) - len(self.TripTripGraph.vs)  
#         num_vehicles = len(self.VehicleTripGraph.vs.select(lambda v: v['id'] < self.num_v))
#         types = [1 if i < num_vehicles else 0 for i in range(len(self.VehicleTripGraph.vs))]
#         self.VehicleTripGraph.vs['type'] = types
        
        number_of_passengers_per_trip = [len(Trips[i]) for i in range(len(Trips))]

        print(Trips)
        self.Trips = Trips
        
        number_of_trips = len(Trips)
        self.num_trips = number_of_trips
#         self.num_vehicles = num_vehicles
        self.num_passen_per_trip = number_of_passengers_per_trip
        
        np.savetxt('number_of_passengers.txt',number_of_passengers_per_trip,fmt = '%d')
        
        np.savetxt('VTGfile.txt',VTAdjacencyMatrix,fmt = '%d')
        
        return Trips
    
    def get_VTM(self):
        return self.VehicleTripMatrix
    
    def get_TTM(self):
        return self.TripTripMatrix
    
    def matrix_to_txt(file,VPAdjacencyMatrix):
        File = np.savetxt(file,VPAdjacencyMatrix,fmt = '%d')
        return File
    
    def write_VTAdjacencyFile(VTAdjacencyMatrix):
        for i in range(len(VTAdjacencyMatrix)):
            for j in range(len(VTAjacencyMatrix[i])):
                if VTAjacencencyMatrix[i][j] == 1:
                    VTAjacencyMatrix[i][j] = 1
                else:
                    VTAjacencyMatrix[i][j] = 0
                    
        np.savetxt('VTGfile.txt',VPAdjacencyMatrix,fmt = '%d')
        
capacity = 4
GraphFile = 'file.txt'
VPFile = 'VPfile.txt'

def matrix_to_txt(file,AdjacencyMatrix):
    np.savetxt(file,AdjacencyMatrix,delimiter = ' ')

class BranchAndBound(VTG):
    def __init__(self,GraphFile,VPFile,capacity):
        vtg = VTG(GraphFile,VPFile,capacity)
        Trips = vtg.VTG()
#         self.VTGraph = vtg.VehicleTripGraph
        self.P = vtg.P
        self.num_vehicles = vtg.num_vehicles
        self.num_trips = vtg.num_trips
        self.trips = vtg.Trips
        self.num_passen_per_trip = vtg.num_passen_per_trip
#         self.TTGraph = vtg.TripTripGraph
        self.Trips = Trips
```

```python
import datetime
import sys

sys.setrecursionlimit(1000000)

BB = BranchAndBound(GraphFile,VPFile,capacity)
```

```python
vehicles = [i for i in range(BB.num_vehicles)] # vehicle 序号
trips = [i for i in range(BB.num_trips)] # trips 序号
num_passen = BB.num_passen_per_trip  # 每个trip中包含的乘客数量
trip_passenger = {i:num_passen[i] for i in trips}  # 将 trip 与 其中的乘客数量对应

with open('VTGfile.txt') as vtf:
    vtg = [[int(num) for num in line.split()] for line in vtf]

#vehicle 是否与trip可匹配， 即图中的边
vehicle_trip = [(i,j) for i in vehicles for j in trips if vtg[i][j] == 1]
# 每条边的权重
w = {(i,j): num_passen[j] for i,j in vehicle_trip} # r in paper
# 每个vehicle可以接待的trip
VT = []
for i in vehicles:
    VT.append([])
    for j in trips:
        if vtg[i][j] == 1:
            VT[i].append(j)
# 每个trip可对应的vehicle
TV = []
for i in trips:
    TV.append([])
    for j in vehicles:
        if vtg[j][i] == 1:
            TV[i].append(j)
            
# passenger 和 trip的对应关系
passen_trips = []

for i in BB.P:
    passen_trips.append([])
    for j in range(len(BB.trips)):
        if i in BB.trips[j]:
            passen_trips[i].append(j)
#         else:
#             passen_trips[i].append(0)
            
# trip 和 passnegers 的关系
trips_passenger = []
for i in range(len(BB.trips)):
    trips_passenger.append([])
    for j in BB.P:
        if j in BB.trips[i]:
            trips_passenger[i].append(j)
            
# oassengers 和  vehicles的关系          
PV = []
# 这里要改
for i in BB.P:
    PV.append([])
    for v in vehicles:
        for j in passen_trips[i]:
            if j in VT[v]:
                PV[i].append(v)
                break

# passengers 和 vheicles的关系
passen_vehicles = []
for p in BB.P:
    passen_vehicles.append([])
    for v in vehicles:
        if v in PV[p]:
            passen_vehicles[p].append(1)
        else:
            passen_vehicles[p].append(0)

# vehicles和passengers的关系
VP = []
for i in vehicles:
    VP.append([])
    for j in BB.P:
        if i in PV[j]:
            VP[i].append(j)

# passengers 和 vehicles的tuple对
passen_vehicles_pair = [(i,j) for i in BB.P for j in vehicles if j in PV[i]]
```

```python
import gurobipy as grp
from gurobipy import *
```

```python
model = grp.Model('Ride Sharing')
```

```python
c_grp = model.addVars(vehicle_trip,obj = vehicle_trip,
                      vtype=grp.GRB.BINARY, name='c_grp')
pv_grp = model.addVars(passen_vehicles_pair,obj = passen_vehicles_pair,
                      vtype = grp.GRB.BINARY, name = 'pvp_grp')
```

```python
model.addConstrs(sum(c_grp[i,j] for j in VT[i])<=1 for i in vehicles)

model.addConstrs(sum(c_grp[i,j] for i in TV[j])<=1 for j in trips)

model.addConstrs(sum(pv_grp[i,j] for j in PV[i])<=1 for i in BB.P)

model.addConstrs(sum(pv_grp[i,j] for i in VP[j])<=capacity for j in vehicles if VP[j]!=[])
```

```python
for j in trips:
    for k in trips_passenger[j]:
        model.addConstr(sum(c_grp[i,j] for i in TV[j])<=sum(pv_grp[k,l] for l in PV[k]))
        
for j in trips:
    for p in trips_passenger[j]:
        for k in trips:
            if (k != j) and (p in trips_passenger[k]):
                model.addConstr(sum(c_grp[i,j] for i in TV[j])+sum(c_grp[s,k] for s in TV[k])<=1)                             
```

```python
model.update()
```

```python
obj = grp.quicksum(w[i,j]*c_grp[i,j] for i,j in vehicle_trip)
model.ModelSense = grp.GRB.MAXIMIZE
```
