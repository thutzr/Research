```python
class VPG(object):
    def __init__(self,GraphFile,VPFile):
        self.graphfile = GraphFile
        self.VPFile = VPFile
        self.VPR = []
        self.PR = []
    
    def readFile(self):
        with open(self.graphfile) as graphfile:
            self.PR = [[int(num) for num in line.split(' ')] for line in graphfile]

        with open(self.VPFile) as VPFile:
            self.VPR = [[int(num) for num in line.split(' ')] for line in VPFile]
            
    
    def PPRelation(self):
        G = dict()
        for i in range(len(self.PR)):
            G[i] = []
            for j in range(len(self.PR)):
                if M[i][j] > 0:
                    G[i] += [j]
        self.PR = G
        return self.PR
    
    def VPRelation(self):
        G = dict()
        for i in range(len(self.VPR)):
            G[i] = []
            for j in range(len(self.VPR[0])):
                if VPR[i][j] > 0:
                    G[i] += [j]
        self.VPR = G
        return self.VPR
    
    def VehicleTripMatch(self,vehicle,Trip):
        if sum([self.VPR[vehicle][trip] == 1 for trip in Trip]) == len(Trip):
            return True
        else:
            return False
    
    def ConstructGraph(self):
        return {'Vehicle-Passenger':self.VPR,'Passenger-Passenger':self.PR}
    
    def Vehicles(self):
        if len(self.VPR) == 0:
            return 0
        else:
            return [i for i in range(len(self.VPR))]
    
    def Passengers(self):
        if len(self.VPR[0]) == 0:
            return 0
        else:
            return [i for i in range(len(self.PR))]
```

```python
class Stack:
    def __init__(self):
        self.items = []
        
    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def clear(self):
        del self.items[:]

    def empty(self):
        return self.size() == 0

    def size(self):
        return len(self.items)

    def top(self):
        return self.items[self.size()-1]
    
class Child(object):
    def __init__(self,VTG,TTG):
        self.VTG = VTG
        self.TTG = TTG

class Node(object):
    def __init__(self, VTGraph,TTGraph,status = 'child',lchild=None, rchild=None):
        self.VTGraph = VTGraph
        self.TTGraph = TTGraph
        self.lchild = lchild
        self.rchild = rchild
        self.w = 0
        self.lw = 0
        self.rw = 0
        self.status = status
 #       self.scc = scc
        
class record(object):
    def __init__(self,Node,state):
        self.node = Node
        self.state = state
     #   self.sccsta = sccsta
```

```python
capacity = 4
GraphFile = 'file.txt'
VPFile = 'VPfile.txt'

def matrix_to_txt(file,AdjacencyMatrix):
    np.savetxt(file,AdjacencyMatrix,delimiter = ' ')

class BranchAndBound(VTG):
    def __init__(self,GraphFile,VPFile,capacity):
        vtg = VTG(GraphFile,VPFile,capacity)
        
        Trips = vtg.VTG()
        self.VTGraph = vtg.VehicleTripGraph
        self.TTGraph = vtg.TripTripGraph
        self.Trips = Trips
        

vtg = VTG(GraphFile,VPFile,capacity)
Trip = vtg.VTG()
num_v = vtg.num_v

def merge_two_dict(x,y):
    z = copy.copy(x)
    z.update(y)
    return z

def MaxDegree(VTGraph,TTGraph):
    '''
    input:
        the trip-trip graph
    output:
        max_deg: the maximum degree
        ind: the id of the trip
    '''
    degrees = TTGraph.vs.degree()
    max_deg = -1
    i = 0
    index = -1
    while i < len(degrees):
        if max_deg < degrees[i]:
            max_deg = degrees[i]
            index = i
        i += 1
    # the index of the vertex in TTG
    ind = TTGraph.vs[index]
    id = ind['id']
    trip_id_in_v = VTGraph.vs.select(lambda v: v['id'] == id)
    
    while len(trip_id_in_v) == 0:
        TTGraph.delete_vertices(index)
        max_deg = TTGraph.maxdegree()
        ind = TTGraph.vs.select(_degree = max_deg)
        index = ind[0].index
        id = ind['id'][0]
        trip_id_in_v = VTGraph.vs.select(lambda v: v['id'] == id)
        
    return (max_deg,id)

def SelectEdge(VTGraph,TTGraph):
    '''
    input:
        VTGraph: the vehicle-trip graph
        TTgraph: the trip-trip graph
    output:
        max_ind: the id of the selected vehicle
    '''
    (deg,ind) = MaxDegree(VTGraph,TTGraph)
    # neighbors receive a vertex's id
  #
  #  print(TTGraph.vs['id'])
  #  print(ind)
    index = VTGraph.vs.select(lambda v: v['id'] == ind)
   # print(index['id'])
    while len(index) == 0:
        remove_ind = TTGraph.vs.select(lambda v: v['id'] == ind)[0].index
        TTGraph.delete_vertices(remove_ind)
        (deg,ind) = MaxDegree(TTGraph)
        index = VTGraph.vs.select(lambda v: v['id'] == ind)
    index = index[0]
  #  print(index)
    neighbors = VTGraph.neighbors(index)
   # print(neighbors)
    if neighbors == []:
        return None
    max_ind = neighbors[0]
    
    max_deg = 0
    for i in range(len(neighbors)):
        if VTGraph.vs[neighbors[i]].degree() > max_deg:
            max_deg = VTGraph.vs[neighbors[i]].degree()
            max_ind = neighbors[i]
            break
    #print(max_ind)
    return (deg,ind,max_ind,TTGraph)

def RemoveEdge(VTGraph,TTGraph,t_id,v_ind):
    '''
    input:
        VTG: the vehicle-trip graph
        TTG: the trip-trip graph
        t_ind: the id of selected trip
    output:
        v_ind: the id of the selected vehicle
        t_ind: the index of the selected trip in TTG
        VTG: the vehicle-trip graph that has an edge deleted
    '''
#    t_ind = TTGraph.vs.select(lambda vertex: vertex['id'] == t_id)[0].index
    t_index = VTGraph.vs.select(lambda v: v['id'] == t_id)[0].index
    VTGraph.delete_edges((v_ind,t_index))
    return (v_ind,t_id,VTGraph,TTGraph)

def RemoveVertex(VTGraph,TTGraph,v_ind,t_id):
    '''
    input:
        VTG
        TTG
        v_ind
        t_id: id of the trip in TTG
    output:
        VTG:
        TTG:
    '''
    # index in TTG of t_ind's neighbors
    t_ind = TTGraph.vs.select(lambda vertex: vertex['id'] == t_id)[0].index
    trip_neighbors = TTGraph.neighbors(t_ind)
    
    trip_id = TTGraph.vs[trip_neighbors]['id']
    trip_index = VTGraph.vs.select(lambda v: v['id'] in trip_id)
    t_index = [v.index for v in trip_index]

#    print(v_ind)
    TTGraph.delete_vertices(trip_neighbors+[t_ind])
  #  print(new_VTG.vs['id'])
  #  print(trip_index+[t_ind,v_ind])
#    t_id = TTGraph.vs['id']
    temp_t_ind = VTGraph.vs.select(lambda v: v['id'] == t_id)[0].index

    VTGraph.delete_vertices(t_index+[temp_t_ind]+[v_ind])
    return (VTGraph,TTGraph)
        
def StronglyConnectedComponents(VTGraph,TTGraph):
    '''
    input:
        VTG
        TTG
    output:
        VTG_scc: list
        TTG_scc: list
    '''
    # SCCs is a VertexClustering Object
    VTG_scc = VTGraph.components().subgraphs()
    TTG_scc = []
    for vtg_cc in VTG_scc:
        scc_id = vtg_cc.vs['id']
        trip_cc = TTGraph.subgraph(TTGraph.vs.select(lambda v: v['id'] in scc_id))
        TTG_scc.append(trip_cc)
    return (VTG_scc,TTG_scc)

def BranchBound(VTGraph,TTGraph):
    s = Stack()
    q = Node(VTGraph = VTGraph,TTGraph = TTGraph,status = 'child')
    state = 0
    while 1:
        VTG = q.VTGraph
        TTG = q.TTGraph
        if q is None:
            if s.empty():
                break
            top = s.top()
            q = top.node
            state = top.state
            s.pop()

        elif len(TTG.vs) == 0:
            if s.empty():
                break
            top = s.top()
            q = top.node
            state = top.state
            s.pop()

        elif len(TTG.vs) == 1:
            if len(VTG.vs) >= 1:
                trip_vs = TTG.vs[0]
                num_passengers = trip_vs['num_passengers']
                # trip is the overall id of the left trip 
                trip = trip_vs['id']   
                # trip_v is the index of the selected trip in VTG
                trip_v = VTG.vs.select(lambda v: v['id'] == trip)[0].index   
                neighbors = VTG.neighbors(trip_v)
                if len(neighbors) == 0:
                    TTG.delete_vertices(trip_vs)
                    VTG.delete_vertices(trip_v)
                    if s.empty():
                        break
                    # back to the former level
                    top = s.top()
                    q = top.node
                    state = top.state
                    s.pop()

                else:
                    # if there is some vehicle left, assign one to the trip
                    vehicle = neighbors[0]
                    if s.empty():
                        break
                    top = s.top()
                    q = top.node
                    state = top.state
                    s.pop()
                    q.w = num_passengers

            else:
                if s.empty():
                    break
                top = s.top()
                q = top.node
                state = top.state
                s.pop()

        elif len(VTG.vs) == 1:
            v = VTG.vs[0]
            v_id = v['id']
            neighbors = VTGraph.neighbors(v)
            if len(neighbors) != 0:
                max_ind = 0
                max_num = 0
                for i in range(len(neighbors)):
                    num = VTGraph.vs[neighbors[i]]['num_passengers']
                    if num > max_num:
                        max_num = num
                        max_ind = i
                id = VTGraph.vs[neighbors[max_ind]]['id']
                if s.empty():
                    break
                top = s.top()
                q = top.node
                state = top.state
                s.pop()
                q.w = max_num

            else:
                if s.empty():
                    break
                top = s.top()
                q = top.node
                state = top.state
                s.pop()

        else:
            if q.status == 'child':
                (vtg_scc,ttg_scc) = StronglyConnectedComponents(VTG,TTG)
                if len(vtg_scc) > 1:
                    for i in range(len(vtg_scc)):
                        s.push(record(Node(VTGraph = vtg_scc[i],TTGraph = ttg_scc[i],status = 'scc'),0))
                    top = s.top()
                    q = top.node
                    state = top.state
                    s.pop()
                else:
                    q.status = 'scc'
            
            _ = SelectEdge(VTG,TTG)
            if _ == None:
                if s.empty():
                    break
                top = s.top()
                q = top.node
                state = top.state
                s.pop()

             #   print(p.VTGraph)
            else:   
                (deg,ind,v_ind,TTG) = _
                if deg == 0:
                    num_vehicles = sum(VTG.vs['type'])
                    MaxMatch = VTG.maximum_bipartite_matching()
                    Match = MaxMatch.matching 
                    Graph = MaxMatch._graph
                    match = dict()
                    num = 0
                    for i in range(num_vehicles):
                        vehicle = Graph.vs[i]['id']
                        if Match[i] != -1:
                            trip = Graph.vs[Match[i]]
                            trip_id = trip['id']
                            trip_num_passengers = trip['num_passengers']
                            match[trip_id] = vehicle
                            num += trip_num_passengers
                    if s.empty():
                        break
                    top = s.top()
                    q = top.node
                    state = top.state
                    s.pop()
                    q.w = num/2

                else:
                    new_VTG = copy.copy(VTG)
                    new_TTG = copy.copy(TTG)
                    (v_id,t_ind,VTG_1,TTG_1) = RemoveEdge(new_VTG,new_TTG,ind,v_ind)
                    if state == 0:
                        s.push(record(q,1))
                        q.lchild = Node(VTGraph = VTG_1,TTGraph = TTG_1,status = 'child')
                        q = q.lchild
                        state = 0

                    elif state == 1:
                        s.push(record(q,2))
                        (VTG_2,TTG_2) = RemoveVertex(new_VTG,new_TTG,v_id,t_ind)
                        q.rchild = Node(VTGraph = VTG_2,TTGraph = TTG_2,status = 'child')
                        q = q.rchild
                        state = 0

                    elif state == 2:
                        if s.empty():
                            break
                        top = s.top()
                        q = top.node
                        state = top.state
                        s.pop()

                        trip = TTG.vs.find(lambda vertex: vertex['id'] == t_ind)
                        if q.rchild == None:
                            q.w = max(q.lchild.w,trip['num_passengers'])
                        else:
                            q.w = max(q.lchild.w,q.rchild.w+trip['num_passengers'])

                            
    print(q.w)
```
