```python
def parser(x):
    return pd.datetime.strptime('190'+x,'%Y-%m')

def train_test_split(df,rate = 0.5):
    train,test = df[:int(df.shape[0]*rate)+1],df[int(df.shape[0]*rate)+1:]
    return (train,test)

def timeseries_to_supervised(data,lag = 1):
    df = pd.DataFrame(data)
   #print(df.shape)
    columns = [df.shift(i) for i in range(1,lag+1)]
    #print(columns)
    columns.append(df)
    df = pd.concat(columns,axis = 1)
    df.fillna(0,inplace = True)
    #print(df.shape)
    return df

def difference(df,interval = 1):
    diff = list()
    for i in range(interval,df.shape[0]):
        diff.append(df[i] - df[i-interval])
    return pd.Series(diff)

def inverse_difference(history,yhat,interval = 1):
    return yhat + history[-interval]

def train_test_scale(train,test):
    scaler = MinMaxScaler(feature_range = (-1,1))
    scaler = scaler.fit(train)
    train = train.reshape(train.shape[0],train.shape[1])
    test = test.reshape(test.shape[0],test.shape[1])
    scaled_train = scaler.transform(train)
    scaled_test = scaler.transform(test)
    
    return scaled_train,scaled_test,scaler

def invert_scale(scaler,X,value):
    new_row = [x for x in X] + [value]
    array = np.array(new_row)
    array = array.reshape(1,len(array))
    inverted = scaler.inverse_transform(array)
    return inverted[0,-1]
    
def fit_lstm_1(train,batch_size,epochs,neurons):
    X, y = train[:, 0:-1], train[:, -1]
    X = X.reshape(X.shape[0],X.shape[1],1)
    inputs = Input(batch_shape = (batch_size,X.shape[1],X.shape[2]))
    lstm_layer_1 = LSTM(neurons,batch_input_shape = (batch_size,X.shape[1],X.shape[2]),
                        activation = 'relu',kernel_regularizer = regularizers.l2(0.1),
                        recurrent_regularizer = regularizers.l2(0.03),
                        bias_regularizer = regularizers.l2(0.01),
                        stateful = True,return_sequences = False)(inputs)
    dense_1 = Dense(neurons,activation = 'relu',name = 'dense_1')(lstm_layer_1)
    added_0 = add([lstm_layer_1,dense_1])
    dense_2 = Dense(neurons,activation = 'relu',name = 'dense_2')(dense_1)
    added_1 = add([dense_2,added_0])
    dense_3 = Dense(neurons,activation = 'relu',name = 'dense_3')(added_1)
    added_2 = add([dense_3,added_1])
    dense_4 = Dense(neurons,activation = 'relu',name = 'dense_4')(added_2)
    added_3 = add([dense_4,added_2])
    dense_5 = Dense(neurons,activation = 'relu',name = 'dense_5')(added_3)
    added_4 = add([dense_5,added_3])
    dense_6 = Dense(neurons,activation = 'relu',name = 'dense_6')(added_4)
    added_5 = add([dense_6,added_4])
    dense_7 = Dense(neurons,activation = 'relu',name = 'dense_7')(added_5)
    added_6 = add([dense_7,added_5])
    dense_8 = Dense(neurons,activation = 'relu',name = 'dense_8')(added_6)
    added_7 = add([dense_8,added_6])
    dense_9 = Dense(neurons,activation = 'tanh',name = 'dense_9')(added_7)
    added_8 = add([dense_9,added_7])
#     dense_10 = Dense(neurons,activation = 'tanh',name = 'dense_10')(added_8)
#     added_9 = add([dense_10,added_8])
#     dense_11 = Dense(neurons,activation = 'relu',name = 'dense_11')(added_9)
#     added_10 = add([dense_11,added_9])
#     dense_12 = Dense(neurons,activation = 'relu',name = 'dense_12')(added_10)
#     added_11 = add([dense_12,added_10])
#     dense_13 = Dense(neurons,activation = 'relu',name = 'dense_13')(added_11)
#     added_12 = add([dense_13,added_11])
#     dense_14 = Dense(neurons,activation = 'relu',name = 'dense_14')(added_12)
#     added_13 = add([dense_14,added_12])
#     dense_15 = Dense(neurons,activation = 'relu',name = 'dense_15')(added_13)
#     added_14 = add([dense_15,added_13])
#     dense_16 = Dense(neurons,activation = 'relu',name = 'dense_16')(added_14)
#     added_15 = add([dense_16,added_14])
#     dense_17 = Dense(neurons,activation = 'relu',name = 'dense_17')(added_15)
#     added_16 = add([dense_17,added_15])
#     dense_18 = Dense(neurons,activation = 'relu',name = 'dense_18')(added_16)
#     added_17 = add([dense_18,added_16])
#     dense_19 = Dense(neurons,activation = 'tanh',name = 'dense_19')(added_17)
#     added_18 = add([dense_19,added_17])
#     dense_20 = Dense(neurons,activation = 'tanh',name = 'dense_20')(added_18)
#     added_19 = add([dense_20,added_18])
    
#     dense_21 = Dense(neurons,activation = 'relu',name = 'dense_21')(added_19)
#     added_20 = add([dense_21,added_19])
#     dense_22 = Dense(neurons,activation = 'relu',name = 'dense_22')(added_20)
#     added_21 = add([dense_22,added_20])
#     dense_23 = Dense(neurons,activation = 'relu',name = 'dense_23')(added_21)
#     added_22 = add([dense_23,added_21])
#     dense_24 = Dense(neurons,activation = 'relu',name = 'dense_24')(added_22)
#     added_23 = add([dense_24,added_22])
#     dense_25 = Dense(neurons,activation = 'relu',name = 'dense_25')(added_23)
#     added_24 = add([dense_25,added_23])
#     dense_26 = Dense(neurons,activation = 'relu',name = 'dense_26')(added_24)
#     added_25 = add([dense_26,added_24])
#     dense_27 = Dense(neurons,activation = 'relu',name = 'dense_27')(added_25)
#     added_26 = add([dense_27,added_25])
#     dense_28 = Dense(neurons,activation = 'relu',name = 'dense_28')(added_26)
#     added_27 = add([dense_28,added_26])
#     dense_29 = Dense(neurons,activation = 'tanh',name = 'dense_29')(added_27)
#     added_28 = add([dense_29,added_27])
#     dense_30 = Dense(neurons,activation = 'tanh',name = 'dense_30')(added_28)
#     added_29 = add([dense_30,added_28])
#     dense_31 = Dense(neurons,activation = 'tanh',name = 'dense_31')(added_29)
#     added_30 = add([dense_31,added_29])
#     dense_32 = Dense(neurons,activation = 'tanh',name = 'dense_32')(added_30)
#     added_31 = add([dense_32,added_30])
    
    
#     dense_33 = Dense(neurons,activation = 'relu',name = 'dense_33')(added_31)
#     added_32 = add([dense_33,added_31])
#     dense_34 = Dense(neurons,activation = 'relu',name = 'dense_34')(added_32)
#     added_33 = add([dense_34,added_32])
#     dense_35 = Dense(neurons,activation = 'relu',name = 'dense_35')(added_33)
#     added_34 = add([dense_35,added_33])
#     dense_36 = Dense(neurons,activation = 'relu',name = 'dense_36')(added_34)
#     added_35 = add([dense_36,added_34])
#     dense_37 = Dense(neurons,activation = 'relu',name = 'dense_37')(added_35)
#     added_36 = add([dense_37,added_35])
#     dense_38 = Dense(neurons,activation = 'relu',name = 'dense_38')(added_36)
#     added_37 = add([dense_38,added_36])
#     dense_39 = Dense(neurons,activation = 'relu',name = 'dense_39')(added_37)
#     added_38 = add([dense_39,added_37])
#     dense_40 = Dense(neurons,activation = 'relu',name = 'dense_40')(added_38)
#     added_39 = add([dense_40,added_38])
#     dense_41 = Dense(neurons,activation = 'tanh',name = 'dense_41')(added_39)
#     added_40 = add([dense_41,added_39])
#     dense_42 = Dense(neurons,activation = 'tanh',name = 'dense_42')(added_40)
#     added_41 = add([dense_42,added_40])
#     dense_43 = Dense(neurons,activation = 'relu',name = 'dense_43')(added_41)
#     added_42 = add([dense_43,added_41])
#     dense_44 = Dense(neurons,activation = 'relu',name = 'dense_44')(added_42)
#     added_43 = add([dense_44,added_42])
#     dense_45 = Dense(neurons,activation = 'relu',name = 'dense_45')(added_43)
#     added_44 = add([dense_45,added_43])
#     dense_46 = Dense(neurons,activation = 'relu',name = 'dense_46')(added_44)
#     added_45 = add([dense_46,added_44])
#     dense_47 = Dense(neurons,activation = 'relu',name = 'dense_47')(added_45)
#     added_46 = add([dense_47,added_45])
#     dense_48 = Dense(neurons,activation = 'relu',name = 'dense_48')(added_46)
#     added_47 = add([dense_48,added_46])
#     dense_49 = Dense(neurons,activation = 'relu',name = 'dense_49')(added_47)
#     added_48 = add([dense_49,added_47])
#     dense_50 = Dense(neurons,activation = 'relu',name = 'dense_50')(added_48)
#     added_49 = add([dense_50,added_48])
#     dense_51 = Dense(neurons,activation = 'tanh',name = 'dense_51')(added_49)
#     added_50 = add([dense_51,added_49])
#     dense_52 = Dense(neurons,activation = 'tanh',name = 'dense_52')(added_50)
#     added_51 = add([dense_52,added_50])
    
#     dense_53 = Dense(neurons,activation = 'relu',name = 'dense_21')(added_19)
#     added_52 = add([dense_21,added_19])
#     dense_54 = Dense(neurons,activation = 'relu',name = 'dense_22')(added_20)
#     added_53 = add([dense_22,added_20])
#     dense_55 = Dense(neurons,activation = 'relu',name = 'dense_23')(added_21)
#     added_54 = add([dense_23,added_21])
#     dense_56 = Dense(neurons,activation = 'relu',name = 'dense_24')(added_22)
#     added_55 = add([dense_24,added_22])
#     dense_57 = Dense(neurons,activation = 'relu',name = 'dense_25')(added_23)
#     added_56 = add([dense_25,added_23])
#     dense_58 = Dense(neurons,activation = 'relu',name = 'dense_26')(added_24)
#     added_57 = add([dense_26,added_24])
#     dense_59 = Dense(neurons,activation = 'relu',name = 'dense_27')(added_25)
#     added_58 = add([dense_27,added_25])
#     dense_60 = Dense(neurons,activation = 'relu',name = 'dense_28')(added_26)
#     added_59 = add([dense_28,added_26])
#     dense_61 = Dense(neurons,activation = 'tanh',name = 'dense_29')(added_27)
#     added_60 = add([dense_29,added_27])
#     dense_62 = Dense(neurons,activation = 'tanh',name = 'dense_30')(added_28)
#     added_61 = add([dense_30,added_28])
#     dense_63 = Dense(neurons,activation = 'tanh',name = 'dense_31')(added_29)
#     added_62 = add([dense_31,added_29])
#     dense_64 = Dense(neurons,activation = 'tanh',name = 'dense_32')(added_30)
#     added_63 = add([dense_32,added_31])
    
    output = Dense(1,name = 'output')(added_8)
    model = Model(inputs = inputs,outputs = output)

    model.compile(optimizer = 'adam',loss = 'mean_squared_error')
    
    try:
        model.load_weights('model_1_weight.h5')
    except:
        pass
    
    historys = list()
    for i in range(epochs):
        history = model.fit(X,y,epochs = 1,batch_size = batch_size,verbose = 0,shuffle = False)
        model.reset_states()
        historys.append(history.history['loss'])
        print('Epoch:'+str(i+1)+' Loss:'+str(history.history['loss']))
        
    model.save_weights('model_1_weight.h5')
    
    return model,historys

def fit_lstm_2(train,batch_size,epochs,neurons):
    X, y = train[:, 0:-1], train[:, -1]
    X = X.reshape(X.shape[0], 1, X.shape[1])
    inputs = Input(batch_shape = (batch_size,X.shape[1],X.shape[2]))
    lstm_layer_1 = LSTM(neurons,batch_input_shape = (batch_size,X.shape[1],X.shape[2]),
                        activation = 'relu',kernel_regularizer = regularizers.l2(0.03),
                        recurrent_regularizer = regularizers.l2(0.03),
                        bias_regularizer = regularizers.l2(0.01),
                        stateful = True,return_sequences = False)(inputs)
    dense_1 = Dense(neurons,activation = 'relu',name = 'dense_1')(lstm_layer_1)
    added_0 = add([lstm_layer_1,dense_1])
    dense_2 = Dense(neurons,activation = 'relu',name = 'dense_2')(added_0)
    dense_3 = Dense(neurons,activation = 'relu',name = 'dense_3')(dense_2)
    added_1 = add([dense_2,dense_3,added_0])
    dense_4 = Dense(neurons,activation = 'relu',name = 'dense_4')(added_1)
    dense_5 = Dense(neurons,activation = 'relu',name = 'dense_5')(dense_4)
    added_2 = add([dense_4,dense_5,added_1])
    dense_6 = Dense(neurons,activation = 'relu',name = 'dense_6')(added_2)
    dense_7 = Dense(neurons,activation = 'relu',name = 'dense_7')(dense_6)
    added_3 = add([dense_6,dense_7,added_2])
    dense_8 = Dense(neurons,activation = 'relu',name = 'dense_8')(added_3)
    dense_9 = Dense(neurons,activation = 'relu',name = 'dense_9')(dense_8)
    added_4 = add([dense_8,dense_9,added_3])
    output = Dense(1,name = 'output')(added_4)
    model = Model(inputs = inputs,outputs = output)
    model.compile(optimizer = 'adam',loss = 'mean_squared_error')
     
    try:
        model.load_weights('model_2_weight.h5')
    except:
        pass
    
    historys = list()
    for i in range(epochs):
        history = model.fit(X,y,epochs = 1,batch_size = batch_size,verbose = 0,shuffle = False)
        model.reset_states()
        historys.append(history.history['loss'])
        print('Epoch:'+str(i+1)+' Loss:'+str(history.history['loss']))
        
    model.save_weights('model_2_weight.h5')
    
    return model,historys

def fit_lstm_3(train,batch_size,epochs,neurons):
    X, y = train[:, 0:-1], train[:, -1]
    X = X.reshape(X.shape[0], 1, X.shape[1])
    #model = Sequential()
    inputs = Input(batch_shape = (batch_size,X.shape[1],X.shape[2]),name = 'input')
    lstm_layer_1 = LSTM(neurons,batch_input_shape = (batch_size,X.shape[1],X.shape[2]),
                        activation = 'relu',kernel_regularizer = regularizers.l2(0.01),
                        recurrent_regularizer = regularizers.l2(0.01),
                        bias_regularizer = regularizers.l2(0.01),
                        stateful = True,return_sequences = True,
                       name = 'lstm_1')(inputs)
    lstm_layer_2 = LSTM(neurons,activation = 'relu',stateful = True,return_sequences = True,name = 'lstm_2')(lstm_layer_1)
    lstm_layer_3 = LSTM(neurons,activation = 'relu',stateful = True,return_sequences = True,name = 'lstm_3')(lstm_layer_2)
    lstm_layer_4 = LSTM(neurons,activation = 'relu',stateful = True,return_sequences = True,name = 'lstm_4')(lstm_layer_3)
    lstm_layer_5 = LSTM(neurons,activation = 'relu',stateful = True,return_sequences = True,name = 'lstm_5')(lstm_layer_4)
    lstm_layer_6 = LSTM(neurons,activation = 'relu',stateful = True,return_sequences = True,name = 'lstm_6')(lstm_layer_5)
    lstm_layer_7 = LSTM(neurons,activation = 'relu',stateful = True,return_sequences = True,name = 'lstm_7')(lstm_layer_6)
    lstm_layer_8 = LSTM(neurons,activation = 'relu',stateful = True,return_sequences = True,name = 'lstm_8')(lstm_layer_7)
    lstm_layer_9 = LSTM(neurons,activation = 'relu',stateful = True,return_sequences = True,name = 'lstm_9')(lstm_layer_8)
    lstm_layer_10 = LSTM(neurons,activation = 'relu',stateful = True,return_sequences = True,name = 'lstm_10')(lstm_layer_9)
    lstm_layer_11 = LSTM(neurons,activation = 'relu',stateful = True,return_sequences = True,name = 'lstm_11')(lstm_layer_10)
    lstm_layer_12 = LSTM(neurons,activation = 'relu',stateful = True,return_sequences = True,name = 'lstm_12')(lstm_layer_11)
    lstm_layer_13 = LSTM(neurons,activation = 'relu',stateful = True,return_sequences = True,name = 'lstm_13')(lstm_layer_12)
    lstm_layer_14 = LSTM(neurons,activation = 'relu',stateful = True,return_sequences = True,name = 'lstm_14')(lstm_layer_13)
    lstm_layer_15 = LSTM(neurons,activation = 'relu',stateful = True,return_sequences = True,name = 'lstm_15')(lstm_layer_14)
    lstm_layer_16 = LSTM(neurons,activation = 'relu',stateful = True,return_sequences = False,name = 'lstm_16')(lstm_layer_15)
    output = Dense(1,name = 'output')(lstm_layer_16)
    model = Model(inputs = inputs,outputs = output)
    model.compile(optimizer = 'adam',loss = 'mean_squared_error')
    
    try:
        model.load_weights('model_2_weight.h5')
    except:
        pass
    
    historys = list()
    for i in range(epochs):
        history = model.fit(X,y,epochs = 1,batch_size = batch_size,verbose = 0,shuffle = False)
        model.reset_states()
        historys.append(history.history['loss'])
        print('Epoch:'+str(i+1)+' Loss:'+str(history.history['loss']))
        
    model.save_weights('model_2_weight.h5')
    
    return model,historys

def fit_lstm_4(train,batch_size,epochs,neurons):
    X, y = train[:, 0:-1], train[:, -1]
    X = X.reshape(X.shape[0], 1, X.shape[1])
    #model = Sequential()
    inputs = Input(batch_shape = (batch_size,X.shape[1],X.shape[2]),name = 'input')
    lstm_layer_1 = LSTM(neurons,batch_input_shape = (batch_size,X.shape[1],X.shape[2]),
                        activation = 'relu',kernel_regularizer = regularizers.l2(0.01),
                        recurrent_regularizer = regularizers.l2(0.01),
                        bias_regularizer = regularizers.l2(0.01),
                        stateful = True,return_sequences = False,
                       name = 'lstm_1')(inputs)
    output = Dense(1,name = 'output')(lstm_layer_1)
    model = Model(inputs = inputs,outputs = output)
    model.compile(optimizer = 'adam',loss = 'mean_squared_error')
    
    try:
        model.load_weights('model_2_weight.h5')
    except:
        pass
    
    historys = list()
    for i in range(epochs):
        history = model.fit(X,y,epochs = 1,batch_size = batch_size,verbose = 0,shuffle = False)
        model.reset_states()
        historys.append(history.history['loss'])
        print('Epoch:'+str(i+1)+' Loss:'+str(history.history['loss']))
        
    model.save_weights('model_2_weight.h5')
    
    return model,historys

def model_5(raw_data):
    mean_value = [0]*144
    MAPE = []
    MAE = []
    RMSE = []
    for i in range(len(raw_data)-144,len(raw_data)):
        j = copy.copy(i)
        mean_value[i-len(raw_data)+144] = list()
        flag = 0
        while(j >= 144):
            mean_value[i-len(raw_data)+144].append(raw_data[j])
            j -= 144
            flag += 1
        mean_value[i-len(raw_data)+144] = np.sum(np.divide(mean_value[i-len(raw_data)+144],flag))
        
    for i in range(144):
#         print(mean_value[i])
        mape = np.divide(np.abs(mean_value[i] - raw_data[i-144]),raw_data[i-144])
        MAPE.append(mape)
        mae = np.absolute(np.subtract(mean_value[i],raw_data[i-144]))
        MAE.append(mae)
        rmse = (mean_value[i]-raw_data[i-144])**2
        RMSE.append(rmse)
#         print(rmse)
    return (np.sqrt(np.sum(RMSE)/144),np.sum(MAPE)/144,np.sum(MAE)/144)
        
def forecast_lstm(model,batch_size,X):
    X = X.reshape(1,1,len(X))
    yhat = model.predict(X,batch_size = batch_size)
    return yhat[0,0]

def plot_loss(historys):
    plt.plot(range(1,len(historys)+1),historys)
    plt.title('Learning Curve')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.show()

def mean_absolute_percentage_error(y_true,y_pred):
    return np.sum(np.divide(np.absolute(np.subtract(y_true,y_pred)),y_true))/len(y_true)

def mean_absolute_error(y_true,y_pred):
    return np.sum(np.absolute(np.subtract(y_true,y_pred)))/len(y_true)

def get_session(gpu_fraction=0.3):

    num_threads = os.environ.get('OMP_NUM_THREADS')
    gpu_options = tf.GPUOptions(per_process_gpu_memory_fraction=gpu_fraction)

    if num_threads:
        return tf.Session(config=tf.ConfigProto(
            gpu_options=gpu_options, intra_op_parallelism_threads=num_threads))
    else:
        return tf.Session(config=tf.ConfigProto(gpu_options=gpu_options))
```
